```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

base = pd.read_csv('sinasc_RO_2019.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
#APGAR1
a = sns.countplot(x= 'APGAR1', data=base)


a.set_title('APGAR1')
a.set_xlabel('Categoria')
a.set_ylabel('Frequência')

plt.show()
```

```python
#APGAR5
plt.close('all')
b = sns.countplot(x= 'APGAR5', data=base)

b.set_title('APGAR5')
b.set_xlabel('Categoria')
b.set_ylabel('Frequência')
plt.show()
```

```python
#SEXO
plt.close('all')


c = sns.countplot(x= 'SEXO', data=base)

c.set_title('Sexo')
c.set_xlabel('')
c.set_ylabel('Frequência')

plt.show()
```

```python
#GESTACAO
plt.close('all')

fig = plt.figure(figsize=(8,5))
                 
d = base['GESTACAO'].value_counts().sort_index().plot.bar()
#d = sns.countplot(x='GESTACAO', data=base)

d.set_title('Período de Gestação')
d.set_xlabel('Semanas de Gestação')
d.set_ylabel('Frequência')


plt.xticks(fontsize=6, rotation=90)
plt.yticks(fontsize=6)

plt.show()
```

```python
#ESTCIVMAE
plt.close('all')
fig = plt.figure(figsize=(8,5))
e = sns.countplot(x= 'ESTCIVMAE', data=base)

e.set_title('Estado Civil da Mãe')
e.set_xlabel('Categoria')
e.set_ylabel('Frequência')


#plt.legend(labels = ['Solteira', 'Casada', 'Viúva', 'Divorciada', 'Ignorado']) cdg não funcionou

plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
f = sns.barplot(x='APGAR5', y='CONSULTAS', data=base)

f.set_title('APGAR5 médio por Consultas de Pré-Natal')
f.set_xlabel('APGAR5')
f.set_ylabel('Consultas')

plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10,8)) # LARGURA, ALTURA

fig, axes = plt.subplots(2,2) # criar figura com os gráficos e cada gráfico no eixo específico como 
# o código abaixo:


#countplot pode ser utilisado para variáveis qualitativas ordinárias

plt.subplots_adjust(wspace=0.3, hspace=0.3) # espaço entre os gráficos

g = sns.countplot(ax= axes[0,0], x= 'SEXO', data=base)

h = sns.countplot(ax= axes[0,1], x= 'CONSULTAS', data=base)

i = sns.countplot(ax= axes[1,0],x= 'ESTCIVMAE', data=base)

j = sns.countplot(ax= axes[1,1], x= 'GESTACAO', data=base)


plt.xticks(fontsize=6, rotation=90)

plt.show()
```
